```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.feature import local_binary_pattern
```

```python
trump = cv2.imread('trump.jpg')
```

```python
trump = cv2.cvtColor(trump, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(trump,cmap = 'gray')
trump.shape
```

```python
# 
hist = np.zeros(256)
rows, cols = trump.shape
for i in range(rows):
    for j in range(cols):
        temp = trump[i][j]
        hist[temp] += 1
```

```python
plt.plot(hist)
plt.show()
```

```python
radius = 1 # LBP
n_points = 8 * radius # 
```

```python
lbp = local_binary_pattern(trump, n_points, radius)
plt.imshow(trump,cmap='gray')
```

```python
plt.imshow(lbp,cmap='gray')
```

```python
# 
```

```python
trans = hist / (rows * cols) * 255
for i in range(1, len(trans)):
    trans[i]=trans[i-1]+trans[i]

gray_h = trump.copy()
for i in range(rows):
    for j in range(cols):
        gray_h[i][j] = int(trans[trump[i][j]])
```

```python
plt.imshow(cv2.hconcat([trump, gray_h]), cmap='gray')
```
